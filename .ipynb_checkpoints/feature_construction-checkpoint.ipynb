```python
import numpy as np
import pandas as  pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('train.csv',usecols=['Age','Pclass','SibSp','Parch','Survived'])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.head()
```

```python
X = df.iloc[:,1:5]
y = df.iloc[:,0]
```

```python
X
```

```python
y
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

**APPLYING FEATURE CONSTRUCTION**

```python
#1 is to add that passenger whose corresponding sibsp and pParch are calculated
X['Family_size']=X['Parch']+X['SibSp']+1
```

```python
X
```

```python
def family_type(num):
    if num==1:
        #alone
        return 0
    elif num>1 and num<=4:
        #small family
        return 1
    else:
        return 2
        #large family
        
```

```python
family_type(8)
```

```python
X['Family_type']=X['Family_size'].apply(family_type)
```

```python
X
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],axis=1)
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```
